```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'Amazon Sales data.csv'
data = pd.read_csv(file_path)
print(data.head())
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
data['Year'] = data['Order Date'].dt.year
data['Month'] = data['Order Date'].dt.month
```

```python
monthly_units_sold = data.groupby('Month')['Units Sold'].sum().reset_index()
monthly_total_revenue = data.groupby('Month')['Total Revenue'].sum().reset_index()
monthly_total_cost = data.groupby('Month')['Total Cost'].sum().reset_index()
monthly_total_profit = data.groupby('Month')['Total Profit'].sum().reset_index()
```

```python
print(monthly_units_sold)
```

```python
print(monthly_total_revenue)
```

```python
print(monthly_total_cost)
```

```python
print(monthly_total_profit)
```

```python
monthly_sales = monthly_units_sold
monthly_sales['Total Revenue'] = monthly_total_revenue['Total Revenue']
monthly_sales['Total Cost'] = monthly_total_cost['Total Cost']
monthly_sales['Total Profit'] = monthly_total_profit['Total Profit']
print(monthly_sales)
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Month',y='Total Revenue',data = monthly_sales,marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.grid(True);
```

```python
year_month_units_sold = data.groupby(['Year', 'Month'])['Units Sold'].sum().reset_index()
year_month_total_revenue = data.groupby(['Year', 'Month'])['Total Revenue'].sum().reset_index()
year_month_total_cost = data.groupby(['Year', 'Month'])['Total Cost'].sum().reset_index()
year_month_total_profit = data.groupby(['Year', 'Month'])['Total Profit'].sum().reset_index()
```

```python
year_month_pivot = data.pivot_table(
    index='Year', columns='Month', values='Total Revenue', aggfunc='sum', fill_value=0
)

# Plotting the heatmap for yearly-monthly sales analysis
plt.figure(figsize=(14, 8))
sns.heatmap(year_month_pivot,fmt=".0f", cmap='YlGnBu', linewidths=0.5, cbar_kws={'label': 'Total Revenue'})
plt.title('Yearly-Monthly Total Revenue Heatmap')
plt.xlabel('Month')
plt.ylabel('Year')
plt.xticks(rotation=0)
plt.yticks(rotation=0)
```

```python
total_units_sold = data['Units Sold'].sum()
total_revenue = data['Total Revenue'].sum()
total_cost = data['Total Cost'].sum()
total_profit = data['Total Profit'].sum()
```

```python
print(f'Total Units Sold: {total_units_sold}')
print(f'Total Revenue: ${total_revenue}')
print(f'Total Cost: ${total_cost}')
print(f'Total Profit: ${total_profit}')
```

```python
# Drop non-numeric columns
columns_to_drop = ['Region', 'Country', 'Item Type', 'Sales Channel', 'Order Priority', 'Order Date', 'Ship Date', 'Order ID']
numeric_data = data.drop(columns=columns_to_drop)

correlation_matrix = numeric_data.corr()

plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix,cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
```
